```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/Sample - Superstore.csv',encoding='ISO-8859-1')
```

```python
data.head(5)
```

```python
# Mengecek jumlah nilai null pada setiap kolom
null_data = data.isnull().sum()

# Menampilkan kolom-kolom dengan nilai null
print(null_data)
```

```python
# Menampilkan baris yang memiliki nilai null
rows_with_null = data[data.isnull().any(axis=1)]

print(rows_with_null)
```

```python

# Memeriksa apakah ada string kosong di kolom 'Profit'
blank_profit = data['Profit'] == ''

# Menampilkan jumlah baris dengan string kosong pada kolom 'Profit'
print("Jumlah baris dengan string kosong di kolom 'Profit':", blank_profit.sum())
```

```python
# Mengonversi kolom 'Order Date' dan 'Ship Date' menjadi tipe data datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], errors='coerce')

# Mengubah format tanggal sehingga bulan menjadi dua digit (MM)
data['Order Date'] = data['Order Date'].dt.strftime('%Y-%m-%d')  # Menjaga format 'YYYY-MM-DD'
data['Ship Date'] = data['Ship Date'].dt.strftime('%Y-%m-%d')    # Menjaga format 'YYYY-MM-DD'

# Menampilkan beberapa baris pertama untuk memeriksa perubahan
print(data[['Order Date', 'Ship Date']].head())
```

```python

# Menyimpan dataset yang telah dimodifikasi ke file Excel
data.to_excel('modified_sample_superstore.xlsx', index=False, engine='openpyxl')

print("File baru telah disimpan dengan nama 'modified_sample_superstore.xlsx'")
```
